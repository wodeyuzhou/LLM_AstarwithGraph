```python
import json
import networkx as nx

def json_to_prompt(graph_json, start, goal, n_points):
    G = nx.node_link_graph(graph_json)

    prompt = "You are a route recommendation expert for a directed transport network.\n\n"

    prompt += "Available Nodes:\n"
    for node in G.nodes:
        prompt += f"- Node ID : {node}\n"

    prompt += "\nAvailable Paths:\n"
    for u, v, attr in G.edges(data=True):
        cost = attr.get('weight', 'Unknown')
        prompt += f"- From {u} to {v} â†’ Cost: {cost}\n"

    prompt += f"\nStart Location: {start}\n"
    prompt += f"Destination: {goal}\n\n"

    prompt += "Task:\n"
    prompt += "1. Find the most efficient route from the start to the destination considering edge costs.\n"
    prompt += f"2. Additionally, recommend {n_points} useful intermediate nodes that would be beneficial to pass through. These could be important connection points, safe areas, or efficient transfer spots.\n\n"

    prompt += "Output Format:\n"
    prompt += "Recommended Waypoints: [$waypolints]\n"

    return prompt
```

```python
with open('graphs/sejong_bus.json', 'r') as f:
    graph_json = json.load(f)


start, goal = 6800031, 6800109
n_points = 1

prompt = json_to_prompt(graph_json, start, goal, n_points)

print(prompt)

with open('prompt.txt', 'w') as f:
    f.write(prompt)
```
